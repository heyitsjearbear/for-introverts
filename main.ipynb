```python
import tensorflow as tf
import numpy as np
# librosa used to load audio files, and porcess them, can do analysis
import librosa
from tensorflow.keras import layers
```

```python
# load audio file test
audio, sr = librosa.load('./dataset/TEST/112.wav',sr=16000)

S = np.abs(librosa.stft(audio, n_fft=512, hop_length=256, win_length=512, window='hann', center=True, pad_mode='reflect'))
print(S)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
img = librosa.display.specshow(librosa.amplitude_to_db(S, ref=np.max), y_axis='log', x_axis='time', ax=ax)
ax.set_title('Power spectrogram')
fig.colorbar(img, ax=ax, format="%+2.0f dB")
```
